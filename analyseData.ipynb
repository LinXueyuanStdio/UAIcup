各字段含义与7月训练数据`train_July.csv`完全相同。

| 字段名称 |  字段含义	 |  数据示例 |

 |  end_geo_id	 |  终点区域id的hash值 | 	 85e49ded1fa70a7bfa01ab0212a6e538 | 
 |  create_date	 |  订单创建日期	 |  2017-08-01 | 
 |  create_hour	 |  订单创建时间（0-23小时）	 |  01 | 
 

此部分数据为public set。

选手在10月16日至1月12日内需要根据这部分数据提交预测结果，并可以在排行榜查看临时成绩。
 

#### `test_id_Aug_agg_private5k.csv`

数据含义和test_Aug_agg_public5k.csv完全相同，但是数据内容与之不同。

此部分数据为private set，该文件在2018-1-12开放下载。

选手在最后3天（即1月12日至1月15日）需要根据这部分数据提交预测结果，最终比赛结果以此结果为准。
 

>关于数据集的其他说明：

>1. 不同的区域id，代表互不重叠的地图区域，起点区域和终点区域有可能相同

### 提交格式 

选手提交的文件须为UTF-8编码(无BOM)的csv格式文件. 数据只需要2列：public set/private set的test_id，和额外加一列并命名为”count”（不含引号），并将预测结果写入这一列。预测结果可以使用浮点数，比如5.3表示比5多一点但不到6. 提交示例如下：

          test_id,count

          0,1

          1,5.3

          ....

 

选手亦可在数据页面下载sample submission作为格式参考.

### 评价指标：

评价指标为MAE(mean absolute error), 即预测值与真实值的绝对误差，并取平均数：
$$ MAE=\frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y_i}| $$
